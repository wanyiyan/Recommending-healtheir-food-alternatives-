```python
import numpy as np
import pandas as pd
```

#### Import Instacart data with nutrition information

```python
df = pd.read_csv('D:/Downloads/Nutrition Data Consolidated Oct_15_with_type_scores_price.csv')
df = df[['food_id', 'name', 'dept_aisle_type_concat', 'dept_aisle_concat', 'SAIN_LIM_Point', 'price', 'rel_price', 'nut_price_ratio', 'nut_rel_price_ratio']]
df.head()
```

#### Recommend products based on nutrition

```python
basket = [1690.0, 1703.0, 16509.0, 16898.0, 26950.0, 36075.0, 40016.0, 56000.0, 56225.0, 56295.0]
```

```python
recommendationdf = []
for i in basket:
    i_name = df.loc[df['food_id'] == i, 'name'].iloc[0]
    three_cat = df.loc[df['food_id'] == i, 'dept_aisle_type_concat'].iloc[0]
    price = df.loc[df['food_id'] == i, 'price'].iloc[0]
    score = df.loc[df['food_id'] == i, 'SAIN_LIM_Point'].iloc[0]
    ratio = df.loc[df['food_id'] == i, 'nut_price_ratio'].iloc[0]
    sub = df[df['dept_aisle_type_concat'] == three_cat]
    slightly_better = (sub[sub['nut_price_ratio'] <= (ratio * 1.2)])
    slightly_better = slightly_better.sort_values(by = ['nut_price_ratio'], ascending = False)
    slightly_better = slightly_better.head(1)
    slightly_id = slightly_better['food_id'].iloc[0]
    slightly_name = slightly_better['name'].iloc[0]
    slightly_price = slightly_better['price'].iloc[0]
    slightly_score = slightly_better['SAIN_LIM_Point'].iloc[0]
    slightly_ratio = slightly_better['nut_price_ratio'].iloc[0]
    medium_better = (sub[sub['nut_price_ratio'] <= (ratio * 1.5)])
    medium_better = medium_better.sort_values(by = ['nut_price_ratio'], ascending = False)
    medium_better = medium_better.head(1)
    medium_id = medium_better['food_id'].iloc[0]
    medium_name = medium_better['name'].iloc[0]
    medium_price = medium_better['price'].iloc[0]
    medium_score = medium_better['SAIN_LIM_Point'].iloc[0]
    medium_ratio = medium_better['nut_price_ratio'].iloc[0]
    best = sub.sort_values(by = ['nut_price_ratio'], ascending = False)
    best = best.head(1)
    best_id = best['food_id'].iloc[0]
    best_name = best['name'].iloc[0]
    best_price = best['price'].iloc[0]
    best_score = best['SAIN_LIM_Point'].iloc[0]
    best_ratio = best['nut_price_ratio'].iloc[0]
    if i == slightly_id:
        two_cat = df.loc[df['food_id'] == i, 'dept_aisle_concat'].iloc[0]
        sub = df[df['dept_aisle_concat'] == two_cat]
        slightly_better = (sub[sub['nut_price_ratio'] <= (ratio * 1.2)])
        slightly_better = slightly_better.sort_values(by = ['nut_price_ratio'], ascending = False)
        slightly_better = slightly_better.head(1)
        slightly_id = slightly_better['food_id'].iloc[0]
        slightly_name = slightly_better['name'].iloc[0]
        slightly_ratio = slightly_better['nut_price_ratio'].iloc[0]
        medium_better = (sub[sub['nut_price_ratio'] <= (ratio * 1.4)])
        medium_better = medium_better.sort_values(by = ['nut_price_ratio'], ascending = False)
        medium_better = medium_better.head(1)
        medium_id = medium_better['food_id'].iloc[0]
        medium_name = medium_better['name'].iloc[0]
        medium_ratio = medium_better['nut_price_ratio'].iloc[0]
        best = sub.sort_values(by = ['nut_price_ratio'], ascending = False)
        best = best.head(1)
        best_id = best['food_id'].iloc[0]
        best_name = best['name'].iloc[0]
        best_ratio = best['nut_price_ratio'].iloc[0]
        output = [i_name, i, ratio, slightly_name, slightly_id, slightly_ratio, medium_name, medium_id, medium_ratio, best_name, best_id, best_ratio]
        recommendationdf.append(output)
    elif i == medium_id:
        two_cat = df.loc[df['food_id'] == i, 'dept_aisle_concat'].iloc[0]
        sub = df[df['dept_aisle_concat'] == two_cat]
        medium_better = (sub[sub['nut_price_ratio'] <= (ratio * 1.5)])
        medium_better = medium_better.sort_values(by = ['nut_price_ratio'], ascending = False)
        medium_better = medium_better.head(1)
        medium_id = medium_better['food_id'].iloc[0]
        medium_name = medium_better['name'].iloc[0]
        medium_ratio = medium_better['nut_price_ratio'].iloc[0]
        best = sub.sort_values(by = ['nut_price_ratio'], ascending = False)
        best = best.head(1)
        best_id = best['food_id'].iloc[0]
        best_name = best['name'].iloc[0]
        best_ratio = best['nut_price_ratio'].iloc[0]
        output = [i_name, i, ratio, slightly_name, slightly_id, slightly_ratio, medium_name, medium_id, medium_ratio, best_name, best_id, best_ratio]
        recommendationdf.append(output)
    elif i == best_id:
        two_cat = df.loc[df['food_id'] == i, 'dept_aisle_concat'].iloc[0]
        sub = df[df['dept_aisle_concat'] == two_cat]
        best = sub.sort_values(by = ['nut_price_ratio'], ascending = False)
        best = best.head(1)
        best_id = best['food_id'].iloc[0]
        best_name = best['name'].iloc[0]
        best_ratio = best['nut_price_ratio'].iloc[0]
        output = [i_name, i, ratio, slightly_name, slightly_id, slightly_ratio, medium_name, medium_id, medium_ratio, best_name, best_id, best_ratio]
        recommendationdf.append(output)
    else:
        output = [i_name, i, price, score, ratio, slightly_name, slightly_id, slightly_price, slightly_score, slightly_ratio, medium_name, medium_id, medium_price, medium_score, medium_ratio, best_name, best_id, best_price, best_score, best_ratio]
        recommendationdf.append(output)
```

```python
column_names = ('original_name', "original_id", "original_price", "original_score", "original_ratio", "economical_name", "economical_id", "economical_price", "economical_score", "economical_ratio", "healthier_name", "healthier_id", "healthier_price", "healthier_score", "healthier_ratio", "healthiest_name", "healthiest_id", "healthiest_price", "healthiest_score", "healthiest_ratio")
recommendationdf = pd.DataFrame(data = recommendationdf, columns = column_names)
```
